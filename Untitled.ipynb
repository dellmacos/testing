```python
from nltk.util import ngrams
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
lis=data[data['Termination Categories [Column V] \n']=='2 - No Scientific Data']['Why Study Stopped']
```

```python
n_grams=[]
words=[]
stop_words=stopwords.words('english')
for sent in lis:
    for word in sent.lower().split():
        if word not in stop_words:
            words.append(word)
for tup in ngrams(words,2):
        n_grams.append(tup)
```

```python
lis[0]
```

```python
from nltk.tag import pos_tag
```

```python
word_pos=[]
for sent in lis:
    pl=pos_tag(sent.split())
    for index in range(len(pl)):
        if pl[index][1]=='JJ':
            if index<len(pl)-1:
                
                word_pos.append(pl[index][0]+' '+pl[index+1][0])
```

```python
dist=FreqDist(word_pos)
```

```python
dist.most_common()
```

```python
dist2=FreqDist(n_grams)
dist2.most_common()
```

```python
n_grams3=[]
words3=[]
stop_words=stopwords.words('english')
for sent in lis:
    for word in sent.lower().split():
        if word not in stop_words:
            words3.append(word)
for tup in ngrams(words3,3):
        n_grams3.append(tup)
```

```python
dist3=FreqDist(n_grams3)
dist3.most_common()
```
